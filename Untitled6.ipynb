```python
from IPython.display import HTML

html_content = """<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>InterHunt Browser</title>
  <meta name="viewport" content="width=device-width, initial-scale=1.0">

  <style>
    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    body, html {
      height: 100%;
      font-family: 'Segoe UI', sans-serif;
      color: white;
      overflow: hidden;
    }

    body {
      background: #000;
      position: relative;
    }

    canvas {
      position: fixed;
      top: 0;
      left: 0;
      z-index: -1;
    }

    .container {
      display: flex;
      flex-direction: column;
      justify-content: center;
      align-items: center;
      height: 100vh;
      padding: 20px;
      text-align: center;
    }

    h1 {
      font-size: 3rem;
      margin-bottom: 10px;
      color: #fff;
      text-shadow: 0 0 10px #f08, 0 0 20px #0ff;
    }

    p {
      margin-bottom: 30px;
      font-size: 1.2rem;
      opacity: 0.85;
    }

    .search-box {
      display: flex;
      background: rgba(255, 255, 255, 0.1);
      border: 1px solid #fff;
      border-radius: 50px;
      padding: 5px;
      box-shadow: 0 0 20px rgba(255, 255, 255, 0.1);
      max-width: 600px;
      width: 100%;
    }

    .search-box input {
      flex: 1;
      padding: 15px 20px;
      font-size: 18px;
      border: none;
      outline: none;
      background: transparent;
      color: white;
    }

    .search-box input::placeholder {
      color: #ccc;
    }

    .search-box button {
      background: #00ffc8;
      border: none;
      padding: 0 20px;
      font-size: 18px;
      cursor: pointer;
      border-radius: 0 50px 50px 0;
      transition: 0.3s;
      color: black;
    }

    .search-box button:hover {
      background: #00cfa5;
    }

    .mic-btn {
      margin-left: 10px;
      background: #ff4081;
      border: none;
      border-radius: 50%;
      color: white;
      width: 50px;
      height: 50px;
      font-size: 22px;
      cursor: pointer;
      animation: glow 2s infinite ease-in-out;
    }

    @keyframes glow {
      0%, 100% {
        box-shadow: 0 0 10px #ff4081;
      }
      50% {
        box-shadow: 0 0 20px #ff4081;
      }
    }

    .footer {
      margin-top: 50px;
      font-size: 14px;
      opacity: 0.6;
    }

    @media (max-width: 600px) {
      h1 {
        font-size: 2rem;
      }

      .search-box input {
        font-size: 16px;
      }

      .search-box button {
        font-size: 16px;
      }
    }
  </style>
</head>
<body>

  <canvas id="bg"></canvas>

  <div class="container">
    <h1>ðŸŽ¨ InterHunt</h1>
    <p>A beautiful blend of art and AI-powered voice search</p>

    <div class="search-box">
      <input type="text" id="searchInput" placeholder="Speak or type your query..." autofocus>
      <button onclick="googleSearch()">Search</button>
      <button class="mic-btn" onclick="startListening()">ðŸŽ¤</button>
    </div>

    <div class="footer">Made for creative minds âœ¨ Show this to your mam proudly!</div>
  </div>

  <script>
    const input = document.getElementById("searchInput");

    input.addEventListener("keydown", function (event) {
      if (event.key === "Enter") {
        googleSearch();
      }
    });

    function googleSearch() {
      const query = input.value.trim();
      if (query) {
        const url = `https://www.google.com/search?q=${encodeURIComponent(query)}`;
        window.open(url, "_blank");
      }
    }

    function startListening() {
      if (!('webkitSpeechRecognition' in window)) {
        alert("Your browser does not support speech recognition.");
        return;
      }

      const recognition = new webkitSpeechRecognition();
      recognition.lang = "en-US";
      recognition.interimResults = false;
      recognition.maxAlternatives = 1;

      input.placeholder = "ðŸŽ§ Listening...";
      recognition.onresult = function (event) {
        const transcript = event.results[0][0].transcript;
        input.value = transcript;
        googleSearch();
      };

      recognition.onerror = function (event) {
        alert("Speech error: " + event.error);
      };

      recognition.onend = function () {
        input.placeholder = "Speak or type your query...";
      };

      recognition.start();
    }
  </script>

  <!-- ðŸŽ¨ Background animation using Canvas API -->
  <script>
    const canvas = document.getElementById("bg");
    const ctx = canvas.getContext("2d");

    let width, height;
    let particles = [];

    function resizeCanvas() {
      width = canvas.width = window.innerWidth;
      height = canvas.height = window.innerHeight;
    }

    window.addEventListener('resize', resizeCanvas);
    resizeCanvas();

    class Particle {
      constructor() {
        this.x = Math.random() * width;
        this.y = Math.random() * height;
        this.radius = Math.random() * 2 + 1;
        this.dx = (Math.random() - 0.5) * 0.5;
        this.dy = (Math.random() - 0.5) * 0.5;
      }

      draw() {
        ctx.beginPath();
        ctx.arc(this.x, this.y, this.radius, 0, Math.PI * 2);
        ctx.fillStyle = "rgba(255, 255, 255, 0.8)";
        ctx.fill();
      }

      update() {
        this.x += this.dx;
        this.y += this.dy;

        if (this.x < 0 || this.x > width) this.dx *= -1;
        if (this.y < 0 || this.y > height) this.dy *= -1;

        this.draw();
      }
    }

    function initParticles(num) {
      particles = [];
      for (let i = 0; i < num; i++) {
        particles.push(new Particle());
      }
    }

    function animate() {
      ctx.clearRect(0, 0, width, height);
      particles.forEach(p => p.update());
      requestAnimationFrame(animate);
    }

    initParticles(100);
    animate();
  </script>

</body>
</html>
"""

display(HTML(html_content))
```
